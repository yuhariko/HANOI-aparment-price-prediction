```python
import csv
from selenium import webdriver
from time import sleep
from bs4 import BeautifulSoup
from unidecode import unidecode
```

```python
def open_browser(url):
    browser = webdriver.Edge()
    browser.get(url)
    return  browser

def insert_data(data,n):
    data.insert(n,str('NaN'))

def Get_URl_Address(browser):
    page_source = BeautifulSoup(browser.page_source)
    # lay toan bo link tu trang
    profiles = page_source.find_all('a', class_ = 'js__product-link-for-product-id')
    all_profile_URL = []
    for profile in profiles:
        profile_ID = profile.get('href')
        profile_URL = 'https://batdongsan.com.vn/' + profile_ID
        if profile_URL not in all_profile_URL:
            all_profile_URL.append(profile_URL)
    #lay dia chi nha cua chung cu
    addresses = page_source.find_all('div', class_ = 're__card-location')
    address_all= []
    for adr in addresses:
        address_all.append(adr.getText().strip())
    return all_profile_URL, address_all

def fill_all_missing_data(data_source, data_pre_fill):
    full_labels = ['Diện tích', 'Mức giá','Hướng nhà', 'Hướng ban công' , 'Số phòng ngủ', 'Số toilet', 'Pháp lý', 'Nội thất']
    data_find = data_source.find_all('span',class_ = 're__pr-specs-content-item-title')
    labels_end = []
    # cac label cua dat
    for i in data_find:
        b = i.getText().strip()
        labels_end.append(b)

    #dien cac label thieu bang NaN
    for i in full_labels:
        if i not in labels_end:
            insert_data(data_pre_fill, full_labels.index(i))

    #dua tat ca ve khong dau
    for j in range(len(data_pre_fill)):
        data_pre_fill[j] = unidecode(data_pre_fill[j])
    return data_pre_fill

def get_place(place):
    plc = ""
    for i in range(45, 78):
        plc += place[i]
    return plc
```

```python
#find and get all Url
#Chon trang bat dau (can thay doi chia nho cac lan ra de giam thoi gian cho)
# Suggest: 150 - 250 trang 1 lan crawl
page_start = 767
#trang ket thuc
page_end = 773

#bat dau lay duong dan
Url = []
Addresses = []
head = 'https://batdongsan.com.vn/ban-can-ho-chung-cu-ha-noi/p'

for i in range(page_start, page_end):
    link_page = head + str(i)
    browser = open_browser(link_page)
    sleep(2)
    Url_, adr_ = Get_URl_Address(browser)
    Url += Url_
    Addresses += adr_
    #chinh thoi gian cho moi lan mo-dong trinh duyet cho phu hop(tuy vao toc do mang)
    #suggest: 5s-20s
    sleep(5)
    browser.close()
```

```python
#mo tung trang de lay thong tin
#thoi gian chay kha lau
dem = 0
trang = 0
full_data = []
for link in Url[27:]:
    dem += 1
    print("dang thuc hien" + str(dem))
    data = open_browser(link)
    data_source = BeautifulSoup(data.page_source)
    item_value = data_source.find_all('span',class_ = 're__pr-specs-content-item-value')
    data_pre_fill = []
    for i in item_value:
        a = i.getText().strip()
        data_pre_fill.append(a)

    #them dia chi vao cuoi du lieu
    if data_source.find('iframe', class_ = 'lazyloaded') is not None:
        place = get_place(data_source.find('iframe', class_ = 'lazyloaded').get('data-src'))
    else:
        place = get_place(data_source.find('iframe', class_ = 'lazyload').get('data-src'))

    data_pre_fill.append(place)
    data_pre_fill.append(Addresses[trang])
    data_pre_fill.append(link)
    trang += 1
    data_after_fill = fill_all_missing_data(data_source,data_pre_fill)
    full_data.append(data_after_fill)

    #dong trang sau moi lan lay du lieu
    sleep(3)
    data.close()
```

```python
#list to Csv
with open('C:\\Users\\huyvq\\PycharmProjects\\pythonProject\\data.csv', 'a') as f:
    writer = csv.writer(f)
    writer.writerows(full_data)
```
